You can compute the mean expression level of a gene in the Crohn's
patients and UC patients, but if they differed, you wouldn't know
whether the difference was real.

One common way to approach this problem, figuring out whether two
means are different, is via a *t-statistic*. The idea is, you look at the
variation in expression levels within Crohn's patients and UC
patients, respectively, to give you an idea of whether the difference
between your means seems big or not. E.g., if the difference between
expression levels is 50, and the standard deviations in Crohn's and UC
patients are 4 and 7, respectively, then that looks like a real
difference because the two distributions don't overlap much.

Okay. Here's the definition of a t-statistic for the difference between
two groups $A$ and $B$:

$$
t = \frac{\mu_{A} - \mu_{B}}{\sqrt{\frac{\sigma^2_{A}}{n_{A}} + \frac{\sigma^2_{B}}{n_{B}}}}
$$

That $\mu$ is the mean of whatever we're measuring for the indicated
group. The $\sigma^2$ is the variance, a standard measure of how
spread-out the data is. The $n$ is the sample size. Here are
definitions for the mean and the variance (this isn't a stats class):

```python
def mean(sample):
	"Compute the average of a list of numbers."
	return sum(sample) / float(len(sample))

def variance(sample):
	"Compute the (sample) variance of a list of numbers."
	mn = mean(sample)
	return sum((n - mn) ** 2 for n in sample) / float(len(sample) - 1)
```

An example: suppose that we had 12 males and 14 females. The mean
height of the males is 6.0 feet and the mean height of the females is
5.7 feet. The variance for the males is 0.0225 and for the females is
0.01 (corresponding to standard deviations of 0.15 feet and 0.1 feet,
respectively). Are the heights of these two groups different, or not?

Plug. Then chug:

$$
t = \frac{\mu_{♂} - \mu_{♀}}{\sqrt{\frac{\sigma^2_{♂}}{n_{♂}} + \frac{\sigma^2_{♀}}{n_{♀}}}} = \frac{6.0 - 5.7}{\sqrt{\frac{0.0225}{12} + \frac{0.01}{14}}} \doteq 5.9
$$

A t-statistic of zero means, naturally, that there's no difference
between the group means. The farther the score gets away from zero,
the more you tend to believe that the means are really different. The
virtue of that fancy denominator is that the units of the t-statistic
are independent of what you measured. A t-statistic of 1 is sort of
unlikely if the means are the same, a t-statistic of 2 is getting
really unlikely that the means are the same, and a t-statistic of 3
means there is almost no chance that the means are the same. A
t-statistic of 5.9 is getting pretty darned close to certainty. So for
these data that's strong evidence that men and women aren't the same
height.
